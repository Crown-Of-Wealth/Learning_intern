```python
# Read data
data = pd.read_csv('Learning_Intern_Raw_Data.csv', sep=',')
data
```

```python
import pandas as pd
import numpy as np
```

```python
# Drop unnecessary columns
df = data[['user_id', 'program', 'assignment_name', 'assignment_type', 'assignment_score', 'has_logged_into_lms', 'Is assignment resubmitted', 'learner_deferred', 'learner_dropped_off', 'overall_score', 'graduated']]
```

```python
# Filter only 'test' assignment types for overall_score
test_scores = df[df['assignment_type'] == 'test']
test_scores
```

```python

# Compute average test score per user
overall_scores = test_scores.groupby('user_id')['assignment_score'].mean()

# Merge computed scores back into the original DataFrame
df.loc[:, 'overall_score'] = df['user_id'].map(overall_scores)

# Save the updated file
df.to_excel("Updated_Learning_Intern.xlsx", index=False)

print("Overall scores updated successfully!")
```

```python
# Define the required milestones for Learner_drop_off
required_milestones = {"Milestone #1", "Milestone #2", "Milestone #3"}

# Filter only required milestones
milestone_scores = df[df['assignment_name'].isin(required_milestones)][['user_id', 'assignment_name', 'assignment_score']]

# Convert scores to numeric (in case they are stored as strings)
milestone_scores['assignment_score'] = pd.to_numeric(milestone_scores['assignment_score'], errors='coerce')

# Identify users who have **all three** milestones with a score greater than 0
valid_users = (
    milestone_scores.groupby('user_id')
    .apply(lambda x: all(x.set_index('assignment_name').loc[list(required_milestones), 'assignment_score'] > 0))
    .reset_index()
)
```

```python
df.info()
```

```python
# Get user IDs who submitted valid milestones
submitted_all = valid_users[valid_users[0]]['user_id']

# Assign learner_dropped_off: 1 (Dropped Off), 0 (Stayed)
df.loc[:, 'learner_dropped_off'] = df['user_id'].apply(lambda x: 0 if x in submitted_all.values else 1)

# Validation Step: Count dropped users
dropped_count = int(df['learner_dropped_off'].sum()/13)
total_users = df['user_id'].nunique()
print(f"Total Learners: {total_users}")
print(f"Learners Dropped Off: {dropped_count}")
print(f"Learners Remaining: {total_users - dropped_count}")

# Save the updated DataFrame
df.to_csv("updated_data.csv", index=False)

print("Users who didn't submit valid milestone scores have been marked as dropped off successfully!")
```

```python
# Ensure Milestones 1 to 5 are submitted
milestones = ["Milestone #1", "Milestone #2", "Milestone #3", "Milestone #4", "Milestone #5"]
user_milestones = df[df["assignment_name"].isin(milestones)].groupby("user_id")["assignment_score"].count()

# Calculate the average score for the first 6 tests
tests = ["Weekly Test #1", "Weekly Test #2", "Weekly Test #3", "Weekly Test #4", "Weekly Test #5", "Weekly Test #6"]
user_test_scores = df[df["assignment_name"].isin(tests)].groupby("user_id")["assignment_score"].mean()
```

```python
# Remove deferred users
df_f = df[df["learner_deferred"] != "Yes"]

# Save the cleaned data
df_f.to_excel("cleaned_dataset.xlsx", index=False)

print("Deferred users removed. Cleaned dataset saved.")
```

```python
# Ensure user-level filtering (not row-based counting)
df_filtered = df.groupby("user_id").filter(
    lambda x: (
        (x["assignment_name"].str.contains("Milestone")).sum() >= 5 and
        (x["assignment_name"].str.contains("Test")).sum() >= 6 and
        (x["assignment_score"].astype(float).mean() >= 75)
    )
)

# Mark unique users as graduated
graduated_users = df_filtered["user_id"].unique()
df["graduated"] = df["user_id"].apply(lambda x: "Yes" if x in graduated_users else "No")

# Count unique graduates
num_graduated = df["user_id"][df["graduated"] == "Yes"].nunique()
print(f"Total number of users who graduated: {num_graduated}")
```

```python
# Function to check if a user qualifies for graduation
def check_graduation(user_df):
    # Get the first 5 milestones
    milestones = user_df[user_df["assignment_name"].str.contains("Milestone")].sort_values("assignment_name").head(5)
    
    # Get all 6 tests
    tests = user_df[user_df["assignment_name"].str.contains("Test")]

    # Check if all first 5 milestones are submitted (i.e., scores > 0)
    if len(milestones) < 5 or (milestones["assignment_score"] == 0).any():
        return "No"

    # Check if user has exactly 6 test scores
    if len(tests) != 6:
        return "No"

    # Check if the average test score is at least 75%
    if tests["assignment_score"].mean() < 75:
        return "No"

    return "Yes"

# Apply function to each user
df["graduated"] = df.groupby("user_id").apply(check_graduation).reset_index(level=0, drop=True)

# Count unique graduates
num_graduated = df["user_id"][df["graduated"] == "Yes"].nunique()
print(f"Total number of users who graduated: {num_graduated}")
```
