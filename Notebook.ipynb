```python
df = data[['user_id', 'program', 'assignment_name', 'assignment_type', 'assignment_score', 'has_logged_into_lms', 'Is assignment resubmitted', 'learner_deferred', 'learner_dropped_off', 'overall_score', 'graduated']]
```

```python
# Filter only 'test' assignment types
test_scores = df[df['assignment_type'] == 'test']
test_scores
```

```python

# Compute average test score per user
overall_scores = test_scores.groupby('user_id')['assignment_score'].mean()

# Merge computed scores back into the original DataFrame
df.loc[:, 'overall_score'] = df['user_id'].map(overall_scores)

# Save the updated file
df.to_excel("Updated_Learning_Intern.xlsx", index=False)

print("Overall scores updated successfully!")
```

```python
# Ensure milestone names are standardized
milestone_names = ["Milestone 1", "Milestone 2", "Milestone 3"]

# Find users who submitted at least one of Milestones 1-3
submitted_milestones = df[df['assignment_name'].isin(milestone_names)]['user_id'].unique()

# Identify users who did NOT submit any of these milestones
all_users = df['user_id'].unique()
dropped_users = set(all_users) - set(submitted_milestones)

# Create a new column and mark dropped users
df['learner_dropped_off'] = df['user_id'].apply(lambda x: 1 if x in dropped_users else 0)

# Save the updated DataFrame
df.to_csv("updated_data.csv", index=False)

print("Learner drop-off status updated successfully!")
```
